```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt

%matplotlib inline
```

```python
df = pd.read_csv("business.retailsales2.csv")
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df.Year.astype(str) + '/' + df.Month.astype(str))
```

```python
plt.figure(figsize=(8,6))
fig = sns.lineplot(data=df, x='Date', y='Gross Sales')
```

```python
df['Date']=df['Date'].map(dt.datetime.toordinal)
```

```python
df.drop(columns=['Month','Year'], inplace=True)
df.dropna(inplace=True)
```

```python
df = pd.DataFrame(df, columns=('Date','Total Orders','Gross Sales','Discounts','Returns','Net Sales','Shipping','Total Sales'))

df
```

```python
df
```

```python
df['Date']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Date']]
X
#,'Total Orders','Gross Sales','Total Sales'
```

```python
y = df[['Net Sales']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=101)
```

```python
X.shape
```

```python
model = LinearRegression()
```

```python
results = model.fit(X_train, y_train)
```

```python
x1 = df.Date.max()
x1_date = df[df.Date.eq(x1)].index[0]
x2 = df.Date.max()
x2_date = df[df.Date.eq(x2)].index[0]

print(x1, x1_date, x2, x2_date)
```

```python
y1 = model.predict(np.array([[x1]]))[0][0]
y2 = model.predict(np.array([[x2]]))[0][0]
```
